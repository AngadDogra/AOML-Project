```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
attendance_data = pd.read_excel("Data/PL DATA.xlsx", "Avg Atten 22 23")
sponsor_data = pd.read_excel("Data/PL DATA.xlsx", "PL Kit Sponsors")

print(attendance_data.head())
print(sponsor_data.head())
```

```python
stadium_to_club = {
    "Old Trafford": "Manchester United",
    "Etihad Stadium": "Manchester City",
    "Anfield": "Liverpool",
    "Stamford Bridge": "Chelsea",
    "Emirates Stadium": "Arsenal",
    "Vitality Stadium": "AFC Bournemouth",
    "Villa Park": "Aston Villa",
    "Brentford Community Stadium": "Brentford",
    "AMEX Stadium": "Brighton",
    "Selhurst Park": "Crystal Palace",
    "Goodison Park": "Everton",
    "Craven Cottage": "Fulham",
    "Elland Road": "Leeds United",
    "King Power Stadium": "Leicester City",
    "St. James's Stadium": "Newcastle United",
    "The City Ground": "Nottingham Forest",
    "St. Mary's Stadium": "Southampton",
    "Tottenham Hotspur Stadium": "Tottenham Hotspur",
    "The London Stadium": "West Ham",
    "Molineux Stadium": "Wolves"
}

attendance_data["Club"] = attendance_data["Stadium"].map(stadium_to_club)

merged_df_harvey_spectre = pd.merge(attendance_data, sponsor_data, on="Club", how="inner")

plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_df_harvey_spectre, x="Sponsor", y="Average")
plt.xticks(rotation=45)
plt.xlabel("Kit Sponsor")
plt.ylabel("Average Attendance")
plt.title("Average Attendance by Kit Sponsor in the Premier League")
plt.show()
```
